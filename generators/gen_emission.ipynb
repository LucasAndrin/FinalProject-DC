```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfco2 = pd.read_csv('../data/generated/gen-co2-emissions.csv')
dfco2[dfco2['Entity'] == 'World']
```

```python
dfghg = pd.read_csv('../data/generated/gen-ghg-emissions.csv')
dfghg[dfghg['Entity'] == 'World']
```

```python
dfmerged = pd.merge(
    dfghg,
    dfco2,
    how='inner',
    on=['Entity', 'Year'],
)
dfmerged[dfmerged['Entity'] == 'World']
```

```python
dfmerged.to_csv(f'../data/generated/gen_emissions ({dfmerged['Year'].min()}-{dfmerged['Year'].max()}).csv', index=False)
dfmerged[dfmerged['Entity'] == 'World'].to_csv('../data/generated/gen-world-emissions.csv')
```
