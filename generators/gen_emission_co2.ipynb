```python
import pandas as pd

# Pré-processamento das emissões de carbono
dfco2 = pd.read_csv('../data/co2-emissions.csv')
dfco2
```

```python
dfco2 = dfco2.drop(columns=['ISO 3166-1 alpha-3'])

# Especificação de colunas
dfco2 = dfco2.rename(columns={
    'Country': 'Entity',
    'Total': 'Total (CO₂)',
    'Gas': 'Gas (CO₂)',
    'Oil': 'Oil (CO₂)',
    'Coal': 'Coal (CO₂)',
    'Other': 'Other (CO₂)',
    'Cement': 'Cement (CO₂)',
    'Flaring': 'Flaring (CO₂)',
    'Per Capita': 'Per Capita (CO₂)',
})

# Ajuste de unidades de medida
co2_columns = [
    'Total (CO₂)',
    'Gas (CO₂)',
    'Oil (CO₂)',
    'Coal (CO₂)',
    'Other (CO₂)',
    'Cement (CO₂)',
    'Flaring (CO₂)',
    'Per Capita (CO₂)',
]

for col in co2_columns:
    if col in dfco2.columns:
        dfco2[col] = dfco2[col] * 1e6

dfco2['Entity'] = dfco2['Entity'].replace('Global', 'World')

dfco2.to_csv('../data/generated/gen-co2-emissions.csv', index=False)
```
