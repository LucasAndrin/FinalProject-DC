```python
import pandas as pd

energy_path = '../data/generated/gen-electricity-sources.csv'
emissions_path = '../data/generated/gen_emissions (1850-2022).csv'

energy = pd.read_csv(energy_path)
emissions = pd.read_csv(emissions_path)

# Filtrar colunas relevantes
emissions_cols = ['Entity', 'Year', 'Total (CO₂)', 'Coal (CO₂)', 'Oil (CO₂)', 'Gas (CO₂)']
energy_cols = ['Entity', 'Year', 'Coal (TWh)', 'Gas (TWh)', 'Oil (TWh)', 'Nuclear (TWh)', 'Hydro (TWh)', 'Solar (TWh)', 'Wind (TWh)']

merged = pd.merge(
    emissions[emissions_cols],
    energy[energy_cols],
    on=['Entity', 'Year'],
    how='inner'
)

merged.to_csv('../data/generated/final-dataset.csv', index=False)
```
